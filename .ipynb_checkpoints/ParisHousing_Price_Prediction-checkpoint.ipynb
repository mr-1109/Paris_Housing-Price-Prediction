# ParisHousing Price Prediction

```python
# import basic libraries
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline
```

```python
# Load the data into dataframe
data=pd.read_csv("ParisHousing.csv")
```

```python
# ckeck the shape of dataframe
data.shape
```

```python
data.info()
```

```python
# check whether dataset has any null values
data.isnull().sum()
```

# Observation:-
1. There is no categorical feature in the dataset.
2. All the features are numerical values

# Now check the correlation between features

```python
plt.figure(figsize=(15,15))
sns.heatmap(data.corr(), cmap="YlGnBu", annot=True)
```

# Observation:-
1. As we can see from the above heatmap, feature 'squarMeters' is highly correlated with feature 'price'

```python
# Distriution plot of numerial data
c=1
plt.figure(figsize=(20,45))

for i in data.columns:
    plt.subplot(6, 3, c)
    plt.xlabel(i)
    sns.distplot(data[i])
    c = c + 1
```

```python
num_col=['squareMeters', 'numberOfRooms', 'hasYard', 'hasPool', 'floors',
       'cityCode', 'cityPartRange', 'numPrevOwners', 'made', 'isNewBuilt',
       'hasStormProtector', 'basement', 'attic', 'garage', 'hasStorageRoom',
       'hasGuestRoom']
```

```python
# Checking the relation between price and all features 
c=1
plt.figure(figsize=(20,45))
for i in num_col:
    plt.subplot(6,3,c)
    sns.scatterplot(x=data[i],y=data['price'])
    plt.xticks(rotation=90)
    plt.tight_layout(pad=3.0)
    c=c+1
```
