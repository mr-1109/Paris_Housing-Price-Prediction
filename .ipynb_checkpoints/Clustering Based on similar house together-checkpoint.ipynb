# ParisHousing Price Prediction

```python
# import basic libraries
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline
```

```python
# Load the data into dataframe
data=pd.read_csv("ParisHousing.csv")
```

```python
data.head()
```

```python
plt.figure(figsize=(15,15))
sns.heatmap(data.corr(), cmap="YlGnBu", annot=True)
```

# Observation:-
1. As we can see from the above heatmap, feature 'squarMeters' is highly correlated with feature 'price'

```python
num_col=[ 'numberOfRooms', 'floors',
       'cityCode', 'cityPartRange', 'numPrevOwners', 'made', 'basement', 'attic', 'garage' ]
```

```python
 x = data.iloc[:,[1,4]]
```

```python
x
```

```python

from sklearn.cluster import KMeans
```

```python
wcss=[]
for i in range(1,10):
    kmeans = KMeans(i)
    kmeans.fit(x)
    wcss_iter = kmeans.inertia_
    wcss.append(wcss_iter)

    number_clusters = range(1,10)
plt.plot(number_clusters,wcss,marker='o')
plt.title('The Elbow title')
plt.xlabel('Number of clusters')
plt.ylabel('WCSS')
```

```python
kmeans = KMeans(5)
kmeans.fit(x)
```

```python
identified_clusters = kmeans.fit_predict(x)
identified_clusters
```

```python
data_with_clusters = data.copy()
data_with_clusters['Clusters'] = identified_clusters 
plt.scatter(data_with_clusters['numberOfRooms'][:100],data_with_clusters['floors'][:100],c=data_with_clusters['Clusters'][:100],cmap='rainbow')
```

```python
wcss=[]
for i in range(1,10):
    kmeans = KMeans(i)
    kmeans.fit(x)
    wcss_iter = kmeans.inertia_
    wcss.append(wcss_iter)

    number_clusters = range(1,10)
plt.plot(number_clusters,wcss)
plt.title('The Elbow title')
plt.xlabel('Number of clusters')
plt.ylabel('WCSS')
```
